```python
import re
import unicodedata
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')

def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

def remove_accented_chars(text):
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    return text

def remove_special_characters(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    lemmatized_text = ' '.join([lemmatizer.lemmatize(token, pos='v') for token in tokens])
    return lemmatized_text
def text_normalizer(text):
    if isinstance(text, str):  # Check if the value is a string
        text = text.lower()
        text = remove_html_tags(text)
        text = remove_accented_chars(text)
        # text = expand_acronyms(text)
        text = remove_special_characters(text)
        text = lemmatize_text(text)
    return text

# Apply text preprocessing to 'reviewText' column
merged_preprocess = mergedone.copy()
merged_preprocess['reviewText'] = merged_preprocess['reviewText'].apply(text_normalizer)

# Display the resulting DataFrame
print(merged_preprocess)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer
from gensim.models import Word2Vec
import numpy as np
# Fill NaN values with an empty string
merged_preprocess['reviewText'].fillna('', inplace=True)

# Bag of Words (BoW) Model
count_vectorizer = CountVectorizer()
bow_features = count_vectorizer.fit_transform(merged_preprocess['reviewText'])


# TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_features = tfidf_vectorizer.fit_transform(merged_preprocess['reviewText'])

# Hashing Vectorizer
hashing_vectorizer = HashingVectorizer()
hashing_features = hashing_vectorizer.fit_transform(merged_preprocess['reviewText'])

# Word2Vec
sentences = [review.split() for review in merged_preprocess['reviewText']]
word2vec_model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)
word2vec_features = []

# Iterate over each review
for review in merged_preprocess['reviewText']:
    # Split the review into words
    words = review.split()
    # Initialize an empty array to store word vectors
    review_vectors = []
    # Iterate over each word in the review
    for word in words:
        # Retrieve the word vector from the Word2Vec model
        if word in word2vec_model.wv:
            review_vectors.append(word2vec_model.wv[word])
    # Compute the mean of all word vectors to obtain the feature vector for the review
    if review_vectors:
        review_vector_mean = np.mean(review_vectors, axis=0)
        word2vec_features.append(review_vector_mean)
    else:
        # If no valid word vectors found (e.g., if all words are out-of-vocabulary), append zeros
        word2vec_features.append(np.zeros(word2vec_model.vector_size))

# Convert the list of feature vectors to a NumPy array
word2vec_features = np.array(word2vec_features)

```

```python
from sklearn.model_selection import train_test_split

# Define a function to convert rating values to classes
def rating_to_class(rating):
    if rating > 3:
        return 'Good'
    elif rating == 3:
        return 'Average'
    else:
        return 'Bad'

# Create a new column 'Rating Class' based on 'overall' ratings
merged_preprocess['Rating Class'] = merged_preprocess['overall'].apply(rating_to_class)

# Split the dataset into input features and target variable
X = merged_preprocess['reviewText']
y = merged_preprocess['Rating Class']

# Split the data into train and test sets (75:25 ratio)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Display the shapes of train and test sets
print("Train set - X:", X_train.shape, " y:", y_train.shape)
print("Test set - X:", X_test.shape, " y:", y_test.shape)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Define pipelines for each model
pipelines = {
    'MultinomialNB': Pipeline([('tfidf', TfidfVectorizer()), ('clf', MultinomialNB())]),
    'Logistic Regression': Pipeline([('tfidf', TfidfVectorizer()), ('clf', LogisticRegression())]),
    'Support Vector Machine': Pipeline([('tfidf', TfidfVectorizer()), ('clf', SVC())]),
    'Decision Tree': Pipeline([('tfidf', TfidfVectorizer()), ('clf', DecisionTreeClassifier())]),
    'Random Forest': Pipeline([('tfidf', TfidfVectorizer()), ('clf', RandomForestClassifier())])
}

# Train and evaluate each model
for model_name, pipeline in pipelines.items():
    print(f"Evaluating {model_name}:")
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    print(classification_report(y_test, y_pred))
    print("------------------------------------------")
```

```python
import pandas as pd

# Define the path to the CSV file
csv_path = '/content/drive/MyDrive/last_df.csv'

# Read the CSV file into a DataFrame
last_df = pd.read_csv(csv_path)

# Display the first few rows of the DataFrame
print(last_df)
```

```python
unique_asins = filtered_df['asin'].unique()

# Filter new_df based on unique_asins
filtered_new_df = last_df[last_df['asin'].isin(unique_asins)]

# Merge filtered rows from both DataFrames
last_merge = pd.merge(filtered_df, filtered_new_df, on='asin', how='inner')

# Display the resulting DataFrame
print(last_merge)
```

```python
import pandas as pd

# Assuming last_df contains reviewerID, asin, and overall ratings
# Aggregate duplicate entries by taking the mean of ratings
user_item_matrix = last_merge.groupby(['reviewerID', 'asin'])['overall'].mean().unstack(fill_value=0)

```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler instance
scaler = MinMaxScaler()

# Normalize the ratings
normalized_ratings = scaler.fit_transform(user_item_matrix)
```

```python
from sklearn.model_selection import KFold
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity

N_values = [10, 20, 30, 40, 50]
K = 5  # Number of folds for K-fold cross-validation

# Initialize list to store MAE values for different values of N
mae_values_user_user = []

# Iterate over different values of N
for N in N_values:
    # Initialize list to store MAE values for each fold
    mae_values_fold = []

    # Perform K-fold cross-validation
    for train_index, val_index in KFold(n_splits=K).split(normalized_ratings):
        # Split data into training and validation sets
        train_set = normalized_ratings[train_index]
        val_set = normalized_ratings[val_index]

        # Compute cosine similarity between users
        user_similarity = cosine_similarity(train_set)

        # Predict ratings for validation set using weighted average of similar users' ratings
        predicted_ratings = np.zeros_like(val_set)
        for i in range(val_set.shape[0]):
            similar_users_indices = np.argsort(user_similarity[i])[-N:]
            similar_users_ratings = train_set[similar_users_indices]
            weights = user_similarity[i][similar_users_indices]
            predicted_ratings[i] = np.sum(similar_users_ratings * weights.reshape(-1, 1), axis=0) / np.sum(weights)

        # Calculate MAE for the fold
        mae = np.mean(np.abs(val_set - predicted_ratings))
        mae_values_fold.append(mae)

    # Calculate average MAE across all folds for the current value of N
    avg_mae = np.mean(mae_values_fold)
    mae_values_user_user.append(avg_mae)
```

```python
# Initialize list to store MAE values for different values of N
mae_values_item_item = []

# Iterate over different values of N
for N in N_values:
    # Initialize list to store MAE values for each fold
    mae_values_fold = []

    # Perform K-fold cross-validation
    for train_index, val_index in KFold(n_splits=K).split(normalized_ratings.T):
        # Split data into training and validation sets
        train_set = normalized_ratings.T[train_index]
        val_set = normalized_ratings.T[val_index]

        # Compute cosine similarity between items
        item_similarity = cosine_similarity(train_set)

        # Predict ratings for validation set using weighted average of similar items' ratings
        predicted_ratings = np.zeros_like(val_set)
        for i in range(val_set.shape[0]):
            similar_items_indices = np.argsort(item_similarity[i])[-N:]
            similar_items_ratings = train_set[similar_items_indices]
            weights = item_similarity[i][similar_items_indices]
            predicted_ratings[i] = np.sum(similar_items_ratings * weights.reshape(-1, 1), axis=0) / np.sum(weights)

        # Calculate MAE for the fold
        mae = np.mean(np.abs(val_set - predicted_ratings))
        mae_values_fold.append(mae)

    # Calculate average MAE across all folds for the current value of N
    avg_mae = np.mean(mae_values_fold)
    mae_values_item_item.append(avg_mae)
```

```python
import matplotlib.pyplot as plt

plt.plot(N_values, mae_values_user_user, label='User-User Recommender')
plt.plot(N_values, mae_values_item_item, label='Item-Item Recommender')
plt.xlabel('N (Number of Similar Users/Items)')
plt.ylabel('MAE (Mean Absolute Error)')
plt.title('MAE vs N for User-User and Item-Item Recommender Systems')
plt.legend()
plt.show()
```
